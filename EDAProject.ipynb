```python
df1=sns.load_dataset('tips')
df1
```

```python
#What is the average tip amount across different days and times?
```

```python
sns.barplot(x="day", y="tip", hue="time", data=df1)
plt.title("Average Tip Amount by Day and Time")
plt.show()
```

```python
#Do male and female customers tip differently?
```

```python
sns.boxplot(x="sex", y="tip", data=df1)
plt.title("Tip Distribution by Gender")
plt.show()
```

```python
#How does the total bill relate to the tip amount — is there a linear trend?
```

```python
sns.regplot(x="total_bill", y="tip", data=df1)
plt.title("Total Bill vs Tip Amount")
plt.show()
```

```python
#Does party size influence the tip percentage?
```

```python
df1["tip_pct"] = df1["tip"] / df1["total_bill"] * 100
sns.boxplot(x="size", y="tip_pct", data=df1)
plt.title("Tip Percentage vs Party Size")
plt.show()
```

```python
#Which day of the week has the highest average total bill?
```

```python
sns.barplot(x="day", y="total_bill", data=df1)
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.boxplot(x="smoker", y="tip", data=df1)
plt.title("Tips by Smoker Status")
plt.show()
```

```python
#Visualize the distribution of total bills and tips using histograms or KDE plots.
```

```python
sns.histplot(df1["total_bill"], kde=True)
plt.title("Distribution of Total Bills")
plt.show()

sns.histplot(df1["tip"], kde=True)
plt.title("Distribution of Tips")
plt.show()
```

```python
df2=sns.load_dataset('iris')
```

```python
df2
```

```python
# Compare the distributions of petal length and width across species.
```

```python
sns.barplot(y="petal_length",x="petal_width",hue="species",data=df2)
plt.title("distributions of petal length and width across species")
plt.show()
```

```python
#Setosa is clearly separated
#Other two species slightly overlap
```

```python
# Which features seem most useful in distinguishing between different species?
```

```python
sns.pairplot(df2, hue="species")
plt.show()
```

```python
#Petal length and petal width are the most useful features for distinguishing between Iris species, 
#as they show clear separation and high variance across species, whereas sepal features exhibit significant overlap
```

```python
# Use pairwise comparisons to identify any strong relationships among features.
```

```python
corr = df2.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```

```python
#Pairwise comparisons show that petal length and petal width have the strongest positive relationship, 
#indicating they grow proportionally and are highly correlated. Sepal width shows weak correlation with other features, 
#making it less informative
```

```python
# Do you notice any clusters or overlaps between the species?
```

```python
sns.scatterplot(x="petal_length",y="petal_width",hue="species",data=df2)
plt.title("Clusters and Overlaps Among Iris Species")
plt.show()
```

```python
#Clear Clusters

#Setosa:
#Forms a tight, well-separated cluster
#Very small petal length and width
#No overlap with other species

# Partial Overlaps
#Versicolor & Virginica
#Overlap in the middle range
#Some points share similar petal sizes
#Not perfectly separable using petal features alone
```

```python
# How does survival rate vary by gender and passenger class?
```

```python
df5=sns.load_dataset('diamonds')
```

```python
df5.head()
```

```python
# How does diamond price vary with carat weight?
```

```python
sns.regplot(data=df5, x='carat', y='price', scatter_kws={'alpha':0.2})
plt.title("Price vs Carat with Trend Line")
plt.show()
```

```python
#Diamond price increases with carat weight
```

```python
# Is there a noticeable price difference between cut categories?
```

```python
sns.boxplot(data=df5, x='cut', y='price')
plt.title("Price Distribution Across Cut Categories")
plt.show()
```

```python
#Yes, there is a noticeable price difference between cut categories, Better cut = higher value, especially when comparing price per carat.
```

```python
# How does clarity and color affect price distribution?
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(data=df5, x='clarity', y='price', hue='color', split=False)
plt.title("Price Distribution by Clarity and Color")
plt.legend(title='Color', bbox_to_anchor=(1.05,1))
plt.show()
```

```python
# Which variables seem most correlated with price?
```

```python
numeric_df = df5.select_dtypes(include='number')

corr = numeric_df.corr()
corr['price'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap (Diamonds Dataset)")
plt.show()
```

```python
#Diamond price is driven mainly by size (carat and dimensions), while quality factors fine-tune the value.
```

```python
# What price range represents most of the dataset (central tendency)?
```

```python
sns.histplot(df5['price'], bins=50, kde=True)
plt.title("Price Distribution of Diamonds")
plt.show()
sns.boxplot(x=df5['price'])
plt.title("Price Spread and Central Range")
plt.show()
```

```python
#A typical diamond in this dataset costs around $2,400, and most diamonds are priced between $1,000 and $5,000.
```

```python
# Does carat weight have a stronger impact on price than clarity or color?
```

```python
df5['price_per_carat'] = df5['price'] / df5['carat']

sns.boxplot(data=df5, x='clarity', y='price_per_carat')
plt.title("Price per Carat by Clarity")
plt.show()
```

```python
df7=sns.load_dataset('car_crashes')
df7
```

```python
# Which states have the highest and lowest accident rates?
```

```python
df7.sort_values(by='total', ascending=False).head(5)[['abbrev', 'total']]
```

```python
df7.sort_values(by='total', ascending=True).head(5)[['abbrev', 'total']]
```

```python
top_states = df7.sort_values('total', ascending=False).head(10)

sns.barplot(data=top_states, x='total', y='abbrev')
plt.title("Top 10 States with Highest Accident Rates")
plt.xlabel("Total Accidents")
plt.ylabel("State")
plt.show()
low_states = df7.sort_values('total').head(10)

sns.barplot(data=low_states, x='total', y='abbrev')
plt.title("Top 10 States with Lowest Accident Rates")
plt.xlabel("Total Accidents")
plt.ylabel("State")
plt.show()
```

```python
# Is there a relationship between speeding and total crashes?
```

```python
df7[['speeding', 'total']].corr()
sns.regplot(data=df7, x='speeding', y='total')
plt.title("Relationship Between Speeding and Total Crashes")
plt.show()
```

```python
#States with higher speeding percentages tend to have more crashes
```

```python
# Does insurance coverage percentage affect fatal crash rates?
```

```python
sns.regplot(data=df7, x='ins_premium', y='total')
plt.title("Insurance Premium vs Fatal Crashes")
plt.show()
df7[['ins_losses', 'total']].corr()
sns.regplot(data=df7, x='ins_losses', y='total')
plt.title("Relationship Between Insurance Losses and Fatal Crashes")
plt.show()
```

```python
#Insurance coverage does not prevent crashes — it manages financial risk after crashes occur.
#Crash rates are more strongly influenced by speeding, alcohol, distractions, and driving behavior.
```

```python
# How do alcohol-related crashes correlate with total crashes?
```

```python
df7[['alcohol', 'total']].corr()
sns.regplot(data=df7, x='alcohol', y='total')
plt.title("Correlation Between Alcohol and Total Crashes")
plt.show()
```

```python
#Alcohol-related crashes show a clear positive correlation with total fatal crashes.
#States with higher alcohol impairment among drivers generally experience more fatal crashes.
```
