```python
### Step 16. What is the average revenue amount per order?
```

```python
data["item_price"]= data["item_price"].replace('[\$,]', '', regex=True).astype(float)
data["revenue"] = data["quantity"] * data["item_price"]
order_totals = data.groupby("order_id")["revenue"].sum()
order_totals.mean()
```
