```python
### How many roles were there in the movie "Inception"?
```

```python
roles=data[data["title"]=="Inception"][["character"]]
len(roles['character'].unique())
```

```python
### How many roles in the movie "Inception" are NOT ranked by an "n" value?
```

```python
roles=data[data["title"]=="Inception"][["n"]].isnull().sum()
roles
```

```python
### But how many roles in the movie "Inception" did receive an "n" value?
```

```python
roles=data[data["title"]=="Inception"][["n"]].notnull().sum()
roles
```

```python
### Display the cast of "North by Northwest" in their correct "n"-value order, ignoring roles that did not earn a numeric "n" value.
```

```python
data[
    (data["title"] == "North by Northwest") &
    (data["n"].notna())
].sort_values(by="n")
```

```python
### Display the entire cast, in "n"-order, of the 1972 film "Sleuth".
```

```python
data[
    (data["title"] == "Sleuth") &
    (data["year"]==1972)&
    (data["n"].notna())
].sort_values(by="n")
```

```python
### Now display the entire cast, in "n"-order, of the 2007 version of "Sleuth".
```

```python
data[
    (data["title"] == "Sleuth") &
    (data["year"]==2007) &
    (data["n"].notna())
].sort_values(by="n")
```
