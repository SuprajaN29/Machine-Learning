```python
#ATM MACHINE SIMULATOR
#Description:
#- Create a program that simulates ATM operations: balance inquiry, cash deposit, cash withdrawal, and PIN verification.
#- Use a dictionary to store user details like {'user_id': {'pin': 1234, 'balance': 10000}}
#- Validate PIN before any transaction.
#- Keep looping until the user exits.
```

```python
users = {
    "user_1": {"pin": 1234, "balance": 10000},
    "user_2": {"pin": 5678, "balance": 5000},
    "user_3": {"pin": 1354, "balance": 50000},
}
def balance_enquiry(user_id):
    print(f'Your balance amount is {users[user_id]['balance']}') 
def cash_deposit(user_id):
    deposit_amount=int(input("Enter amount to be deposited: "))
    users[user_id]["balance"] += deposit_amount
    print(f'{deposit_amount} deposited successfully')
    print(f'Updated Balance: {users[user_id]['balance']} ')
def cash_withdrwal(user_id):
    withdraw_amount=int(input("Enter amount to be withdrawn: "))
    if withdraw_amount>users[user_id]['balance']:
        print("Insufficient Balance")
    else:
         users[user_id]["balance"] -= withdraw_amount
         print("Amount withdrawn Successfully")
         print(f'Updated Balance: {users[user_id]['balance']} ')       
def showMenu(user_id):
    while True:
        print("\n--- ATM MENU ---")
        print("1. Balance Inquiry")
        print("2. Cash Deposit")
        print("3. Cash Withdrawal")
        print("4. Exit")
        choice = input("Choose an option (1-4): ")

        if choice == "1":
            balance_enquiry(user_id)

        elif choice == "2":
           cash_deposit(user_id)

        elif choice == "3":
            cash_withdrwal(user_id)

        elif choice == "4":
            print("Thank you for using the ATM ")
            break

        else:
            print("Invalid choice, Please try again.")
    
def pin_verificaion(user_id):
    pin=int(input("Please enter your pin: "))
    if pin==users[user_id]["pin"]:
        print("login Success")
        showMenu(user_id)
    else:
        print("Incorrect pin, Try Again Later")
def atm_simulator():
    user_id=input("Enter User Id: ")
    if user_id not in users:
        print("User not found")
    else:
        pin_verificaion(user_id)

atm_simulator()
```

```python
#STUDENT GRADE MANAGEMENT SYSTEM
#Description:
#- Allow the user to add student names, subjects, and marks.
#- Calculate total, average, and highest scorer.
#- Store each student’s record in a dictionary.
#- Display class average and topper.
```

```python
student_records = {}
student_id = 0

def add_record():
    global student_id

    name = input("Enter Student Name: ")
    subjects = int(input("Enter number of subjects: "))
    marks = {}

    for i in range(subjects):
        subject_name = input("Enter Subject Name: ")
        score = int(input(f"Enter marks scored in {subject_name}: "))
        marks[subject_name] = score

    total = sum(v for v in marks.values())
    average = total / subjects

    student_records[student_id] = {
        "Name": name,
        "Marks": marks,
        "Total": total,
        "Average": average
    }

    print("Student record added successfully ")
    print(f"Student ID: {student_id}")

    student_id += 1


while True:
    print("\n.....Menu....")
    print("1. Add record")
    print("2. Display record")
    print("3. Highest Scorer")
    print("4. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        add_record()

    elif choice == "2":
        student = int(input("Enter student Id: "))
        if student in student_records:
            print(student_records[student])
        else:
            print("Student data not found ")

    elif choice == "3":
        if not student_records:
            print("No records available ")
        else:
            topper = max(student_records, key=lambda x: student_records[x]["Total"])
            print("Highest Scorer:")
            print(f"Student ID: {topper}")
            print(student_records[topper])

    elif choice == "4":
        
        print("Exiting program ")
        break

    else:
        print("Invalid choice")
```

```python
#CONTACT BOOK
#Description:
#- Add, update, delete, and search contacts.
#- Store name, phone, and email in a dictionary.
#- Display all contacts sorted alphabetically.
```

```python
#LIBRARY MANAGEMENT SYSTEM
#Description:
#- Maintain a list of books.
#- Features: Add new books, issue books, return books, search by title/author.
#- Track available and issued books.
```

```python
library = {}

def add_book():
    book_id = input("Enter Book ID: ")
    title = input("Enter Book Title: ")
    author = input("Enter Author Name: ")
    copies = int(input("Enter number of copies: "))

    library[book_id] = {
        "title": title,
        "author": author,
        "copies": copies,
        "issued": 0
    }
    print("Book added successfully!")

def issue_book():
    book_id = input("Enter Book ID to issue: ")

    if book_id in library:
        if library[book_id]["copies"] > 0:
            library[book_id]["copies"] -= 1
            library[book_id]["issued"] += 1
            print("Book issued successfully!")
        else:
            print("No copies available!")
    else:
        print("Book not found!")

def return_book():
    book_id = input("Enter Book ID to return: ")

    if book_id in library and library[book_id]["issued"] > 0:
        library[book_id]["copies"] += 1
        library[book_id]["issued"] -= 1
        print("Book returned successfully!")
    else:
        print("Invalid return!")

def search_book():
    keyword = input("Enter title or author to search: ").lower()
    found = False

    for book in library.values():
        if keyword in book["title"].lower() or keyword in book["author"].lower():
            print(book)
            found = True

    if not found:
        print("No matching book found!")

def display_books():
    if not library:
        print("Library is empty!")
        return

    for book_id, book in library.items():
        print(f"\nBook ID: {book_id}")
        print(f"Title: {book['title']}")
        print(f"Author: {book['author']}")
        print(f"Available: {book['copies']}")
        print(f"Issued: {book['issued']}")

# Main menu
while True:
    print("\n Library Management System")
    print("1. Add Book")
    print("2. Issue Book")
    print("3. Return Book")
    print("4. Search Book")
    print("5. Display All Books")
    print("6. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        add_book()
    elif choice == "2":
        issue_book()
    elif choice == "3":
        return_book()
    elif choice == "4":
        search_book()
    elif choice == "5":
        display_books()
    elif choice == "6":
        print("Exiting Library System. Goodbye!")
        break
    else:
        print("Invalid choice. Try again.")
```

```python
#SIMPLE BANKING LEDGER
#Description:
#- Record transactions (date, type, amount).
#- Allow users to view transaction history and total balance.
```

```python
ledger = []

def add_transaction():
    date = input("Enter date (DD-MM-YYYY): ")
    trans_type = input("Enter type (credit/debit): ").lower()
    amount = float(input("Enter amount: "))

    if trans_type not in ["credit", "debit"]:
        print("Invalid transaction type!")
        return

    ledger.append({
        "date": date,
        "type": trans_type,
        "amount": amount
    })

    print("Transaction recorded successfully!")

def view_transactions():
    if not ledger:
        print("No transactions found!")
        return

    print("\n Transaction History:")
    for t in ledger:
        print(f"{t['date']} | {t['type'].capitalize()} | ₹{t['amount']}")

def total_balance():
    balance = 0
    for t in ledger:
        if t["type"] == "credit":
            balance += t["amount"]
        else:
            balance -= t["amount"]

    print(f"\n Total Balance: ₹{balance}")

# Main menu
while True:
    print("\n--- Banking Ledger Menu ---")
    print("1. Add Transaction")
    print("2. View Transactions")
    print("3. View Balance")
    print("4. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        add_transaction()
    elif choice == "2":
        view_transactions()
    elif choice == "3":
        total_balance()
    elif choice == "4":
        print("Exiting Banking Ledger. Goodbye!")
        break
    else:
        print("Invalid choice! Try again.")
```

```python
#INVENTORY MANAGEMENT SYSTEM
#Description:
#- Manage product inventory with fields: name, quantity, and price.
#- Add new products, update quantities, calculate total value.
```

```python
inventory = {}

def add_product():
    name = input("Enter product name: ").title()
    quantity = int(input("Enter quantity: "))
    price = float(input("Enter price per unit: "))

    if name in inventory:
        print("Product already exists!")
    else:
        inventory[name] = {
            "quantity": quantity,
            "price": price
        }
        print("Product added successfully!")

def update_quantity():
    name = input("Enter product name to update: ").title()

    if name in inventory:
        quantity = int(input("Enter new quantity: "))
        inventory[name]["quantity"] = quantity
        print("Quantity updated successfully!")
    else:
        print("Product not found!")

def total_inventory_value():
    total_value = sum(
        item["quantity"] * item["price"]
        for item in inventory.values()
    )
    print(f"\n Total Inventory Value: ₹{total_value}")

def display_inventory():
    if not inventory:
        print("Inventory is empty!")
        return

    print("\n Inventory List:")
    for name, item in inventory.items():
        print(f"{name} → Qty: {item['quantity']}, Price: ₹{item['price']}")

# Main menu
while True:
    print("\n--- Inventory Management Menu ---")
    print("1. Add Product")
    print("2. Update Quantity")
    print("3. Total Inventory Value")
    print("4. Display Inventory")
    print("5. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        add_product()
    elif choice == "2":
        update_quantity()
    elif choice == "3":
        total_inventory_value()
    elif choice == "4":
        display_inventory()
    elif choice == "5":
        print("Exiting Inventory Management System. Goodbye!")
        break
    else:
        print("Invalid choice! Try again.")
```

```python
#VOTING SYSTEM
#Description:
#- Register candidates and voters.
#- Allow one vote per voter.
#- Count and display votes with the winner.
```

```python
candidates = {}
voters = {}

def register_candidate():
    name = input("Enter candidate name: ")
    if name in candidates:
        print("Candidate already registered")
    else:
        candidates[name] = 0
        print("Candidate registered successfully")

def register_voter():
    voter_id = input("Enter voter ID: ")
    if voter_id in voters:
        print("Voter already registered")
    else:
        voters[voter_id] = False   # False means not voted
        print("Voter registered successfully")

def cast_vote():
    voter_id = input("Enter your voter ID: ")

    if voter_id not in voters:
        print("Voter not registered")
        return

    if voters[voter_id]:
        print("You have already voted")
        return

    print("\nCandidates:")
    for name in candidates:
        print("-", name)

    choice = input("Enter candidate name to vote: ")

    if choice in candidates:
        candidates[choice] += 1
        voters[voter_id] = True
        print("Vote cast successfully")
    else:
        print("Invalid candidate")

def show_results():
    print("\n VOTING RESULTS")
    for name, votes in candidates.items():
        print(f"{name}: {votes} votes")

    if candidates:
        winner = max(candidates, key=candidates.get)
        print(f"\n Winner: {winner}")
    else:
        print("No candidates available")

while True:
    print("\n--- VOTING SYSTEM MENU ---")
    print("1. Register Candidate")
    print("2. Register Voter")
    print("3. Cast Vote")
    print("4. Show Results")
    print("5. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        register_candidate()
    elif choice == "2":
        register_voter()
    elif choice == "3":
        cast_vote()
    elif choice == "4":
        show_results()
    elif choice == "5":
        print("Exiting Voting System")
        break
    else:
        print("Invalid choice")
```

```python
#MOVIE TICKET BOOKING SYSTEM
#Description:
#- Represent seats in a 2D list.
#- Book seats and mark them as 'X'.
#- Calculate ticket prices based on seat type.
```

```python
# Seat layout (5 rows x 5 columns)
# F = Front (₹150), M = Middle (₹200), B = Back (₹250)

seats = [
    ['F', 'F', 'F', 'F', 'F'],
    ['M', 'M', 'M', 'M', 'M'],
    ['M', 'M', 'M', 'M', 'M'],
    ['B', 'B', 'B', 'B', 'B'],
    ['B', 'B', 'B', 'B', 'B']
]

prices = {'F': 150, 'M': 200, 'B': 250}

def show_seats():
    print("\n Seat Layout (X = Booked)")
    for row in seats:
        print(" ".join(row))

def book_seat():
    row = int(input("Enter row number (0-4): "))
    col = int(input("Enter seat number (0-4): "))

    if row < 0 or row >= 5 or col < 0 or col >= 5:
        print("Invalid seat position")
        return

    if seats[row][col] == 'X':
        print("Seat already booked")
        return

    seat_type = seats[row][col]
    price = prices[seat_type]

    seats[row][col] = 'X'
    print(f"seat booked successfully | Price: ₹{price}")

while True:
    print("\n--- MOVIE TICKET BOOKING MENU ---")
    print("1. Show Seats")
    print("2. Book Seat")
    print("3. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        show_seats()
    elif choice == "2":
        book_seat()
    elif choice == "3":
        print("Enjoy the movie!")
        break
    else:
        print("Invalid choice")
```

```python
#EXPENSE TRACKER
#Description:
#- Track daily expenses by category, amount, and date.
#- Calculate total per category and monthly summary.
```

```python
expenses = []
def add_expense():
    date = input("Enter date (YYYY-MM-DD): ")
    category = input("Enter category (Food, Travel, Rent, etc.): ")
    amount = float(input("Enter amount: "))

    expenses.append({
        "date": date,
        "category": category,
        "amount": amount
    })

    print("Expense added successfully")
    
def view_expenses():
    if not expenses:
        print("No expenses recorded")
        return

    print("\n ALL EXPENSES")
    for exp in expenses:
        print(exp)

def total_by_category():
    category_totals = {}

    for exp in expenses:
        cat = exp["category"]
        category_totals[cat] = category_totals.get(cat, 0) + exp["amount"]

    print("\n TOTAL PER CATEGORY")
    for cat, total in category_totals.items():
        print(f"{cat}: ₹{total}")

def monthly_summary():
    month = input("Enter month (YYYY-MM): ")
    total = 0

    for exp in expenses:
        if exp["date"].startswith(month):
            total += exp["amount"]

    print(f"\n Total expenses for {month}: ₹{total}")

while True:
    print("\n--- EXPENSE TRACKER MENU ---")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Total by Category")
    print("4. Monthly Summary")
    print("5. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expenses()
    elif choice == "3":
        total_by_category()
    elif choice == "4":
        monthly_summary()
    elif choice == "5":
        print("Exiting Expense Tracker")
        break
    else:
        print("Invalid choice")
```

```python
#BANKING CUSTOMER DATABASE
#Description:
#- Maintain customer data with fields: ID, name, type, balance.
#- Perform deposit, withdrawal, and account summary operations.
```

```python
customers = {}

def add_customer():
    cust_id = input("Enter Customer ID: ")
    name = input("Enter Customer Name: ")
    acc_type = input("Enter Account Type (Savings/Current): ")
    balance = float(input("Enter Initial Balance: "))

    customers[cust_id] = {
        "Name": name,
        "Type": acc_type,
        "Balance": balance
    }

    print("Customer added successfully")

def deposit():
    cust_id = input("Enter Customer ID: ")
    if cust_id in customers:
        amount = float(input("Enter amount to deposit: "))
        customers[cust_id]["Balance"] += amount
        print("Amount deposited successfully")
    else:
        print("Customer not found")

def withdraw():
    cust_id = input("Enter Customer ID: ")
    if cust_id in customers:
        amount = float(input("Enter amount to withdraw: "))
        if amount > customers[cust_id]["Balance"]:
            print("Insufficient balance")
        else:
            customers[cust_id]["Balance"] -= amount
            print("Withdrawal successful")
    else:
        print("Customer not found")

def account_summary():
    cust_id = input("Enter Customer ID: ")
    if cust_id in customers:
        print("\nACCOUNT SUMMARY")
        for key, value in customers[cust_id].items():
            print(f"{key}: {value}")
    else:
        print("Customer not found")

while True:
    print("\n--- BANKING SYSTEM MENU ---")
    print("1. Add Customer")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Account Summary")
    print("5. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        add_customer()
    elif choice == "2":
        deposit()
    elif choice == "3":
        withdraw()
    elif choice == "4":
        account_summary()
    elif choice == "5":
        print("Exiting Banking System")
        break
    else:
        print("Invalid choice")
```

```python
#Even and Odd Filter:
    #- Given a list of integers, use list comprehension to separate even and odd numbers.
```

```python
integers=[11,34,56,77,58,33,67,13,68,88,42]
even_numbers=[i for i in integers if i%2==0]
odd_numbers=[i for i in integers if i%2!=0]
print(even_numbers)
print(odd_numbers)
```

```python
#Vowel Counter:
   # - Input a string and count the number of vowels using loops.
```

```python
str=input("enter a string: ")
count=0
vowels=['a','e','i','o','u','A','E','I','O','U']
for i in list(str):
    if i in vowels:
        count+=1
print(count)
```

```python
#Common Elements Finder:
   # - Find common elements between two lists using list comprehension.
```

```python
list_one=[23,46,76,323,657,32,677,435,34,45]
list_two=[545,65,34,23,67,89,78]
common_elements=[i for i in set(list_one).intersection(set(list_two))]
print(list(common_elements))
```

```python
# Dictionary from Two Lists:
    #- Create a dictionary by zipping two lists (keys and values).
```

```python
# Word Length Mapping:
   # - Create a dictionary where keys are words and values are their lengths using comprehension
```

```python
words=["ram","sita","hanuman","shiva","parvathi"]
my_dict={word: len(word) for word in words}
print(my_dict)
```

```python
#Prime Numbers in a Range:
    #- Generate all prime numbers between 1 and 100 using list comprehension
```

```python
prime_numbers = [
    i for i in range(1, 101)
    if all(i % j != 0 for j in range(2, int(i**0.5) + 1))
]
print(prime_numbers)
```

```python
#Flatten a Nested List:
    #- Convert [[1,2],[3,4],[5,6]] into [1,2,3,4,5,6] using nested list comprehension.
```

```python
my_list=[[1,2],[3,4],[5,6]]
new_list = [j for i in my_list for j in i]
print(new_list)
```
